**Here we will check, whether the neural networks implemented in nn.ipynb works fine**

```python
from NeuralNetwork import NN, GD
import numpy as np
import matplotlib.pyplot as plt
```

Tests on simple example (learning 1)

```python
layers = [1, 2, 3, 2, 1]    # give wanted structure of neural nettwork

nn = NN(layers)
optimizer = GD(lr=1)

X = np.array([[1]]) # input has to be numpy matrix
y = [1]

iterations = 1000
losses = []

print("Before training: ",nn.forward(X))

for i in range(iterations):
    optimizer.step(nn, X, y)

    y_pred = nn.forward(X)
    loss = (y_pred - y) ** 2
    losses.append(loss)

print("After training: ", nn.forward(X))

# plot losses over iterations
flat_losses = [loss[0] if isinstance(loss, np.ndarray) else loss for loss in losses]  # model returns matrix output, so we have to flatten it
plt.plot(flat_losses)
```

looks good, the error over iterations is going to 0
